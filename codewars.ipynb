Given a non-empty array of integers, return the result of multiplying the values together in order

```python
def grow(arr):
    size = len(arr)
    if(size > 0):
        result = 1
        for num in arr:
            result = result * num
        return result
    else:
        pass

grow([1, 2, 3, 4])
```
